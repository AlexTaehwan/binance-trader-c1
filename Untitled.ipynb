```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(3, 1, figsize=(12, 9))

pd.Series(np.random.randn(1000)).plot(ax=ax[0])
pd.Series(np.random.randn(1000)).plot(ax=ax[1])
pd.Series(np.random.randn(1000)).plot(ax=ax[2])

ax[0].set_title('a')

plt.tight_layout()
plt.show()
```
